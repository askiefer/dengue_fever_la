```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('../data/dengue_features_entire.csv')
```

Delete the rows with NaNs

```python
df = df.dropna()
```

Separate the data for each city

```python
sj = df[df['city'] == "sj"]
```

```python
iq = df[df['city'] == "iq"]
```

```python
len(sj), len(iq)
```

```python
sj.head()
```

```python
plt.scatter(sj['year'], sj['total_cases'])
plt.xlabel('year')
plt.ylabel('total cases')
plt.suptitle('Year vs. Total Cases')
```

```python
plt.scatter(sj['weekofyear'], sj['total_cases'])
plt.xlabel('week of year')
plt.ylabel('total cases')
plt.suptitle('week of year vs. Total Cases')
```

```python
# removed reanalysis_sat_precip_amt_mm as column is
# identical to precipitation_amt_mm
columns = ['year',
 'weekofyear',
 'ndvi_ne',
 'ndvi_nw',
 'ndvi_se',
 'ndvi_sw',
 'precipitation_amt_mm',
 'reanalysis_air_temp_k',
 'reanalysis_avg_temp_k',
 'reanalysis_dew_point_temp_k',
 'reanalysis_max_air_temp_k',
 'reanalysis_min_air_temp_k',
 'reanalysis_precip_amt_kg_per_m2',
 'reanalysis_relative_humidity_percent',
 'reanalysis_specific_humidity_g_per_kg',
 'reanalysis_tdtr_k',
 'station_avg_temp_c',
 'station_diur_temp_rng_c',
 'station_max_temp_c',
 'station_min_temp_c',
 'station_precip_mm']
for column in columns:
    plt.scatter(sj[column], sj['total_cases'])
    plt.xlabel(column)
    plt.ylabel('total cases')
    title = '{}'.format(column).replace("_", " ").title()
    plt.suptitle(title + ' vs. Total Cases')
    plt.show()
```

## Variable correlations with total cases:
- precipitation_amt_mm - weak negative

- reanalysis_dew_point_k - weak positive

- reanalysis_min_air_temp_k - weak positive
- reanalysis_relative_humidity_percent - weak positive
- reanalysis_specific_humidity_g_per_kg - weak positive
- station_avg_temp - positive
